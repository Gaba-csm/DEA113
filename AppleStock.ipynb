```python
import pandas as pd
url = 'https://raw.githubusercontent.com/Gaba-csm/DEA113/refs/heads/main/apple_stock.csv'
data = pd.read_csv(url)
data.head()
```

```python
# Calculate the difference between High and Close and store it in a new column "DifferenceOfHighandClose"
data['DifferenceOfHighandClose'] = data['High'] - data['Close']
data.head()
```

```python
data.shape
```

```python
#REMOVING THE NEW COLUMN COLUMN axis 1=column, 0=row
data = data.drop(['DifferenceOfHighandClose'], axis = 1)
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
print("Mean of Adj Close is:", data ['Adj Close'].mean())
print("Median of Adj Close is:", data ['Adj Close'].median())
print("Mode of Adj Close is:", data ['Adj Close'].mode())
```

```python
data.rename(columns={'Unnamed: 0': 'Date'}, inplace=True)
data.head()
```

```python
data ['Date'] = pd.to_datetime(data['Date'])
data['year'] = data['Date'].dt.year
Average_per_year = data.groupby('year')['Adj Close'].mean()
print(Average_per_year)
```
